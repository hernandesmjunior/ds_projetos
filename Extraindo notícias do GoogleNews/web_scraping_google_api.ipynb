```python
from GoogleNews import GoogleNews
from newspaper import Article
from newspaper import Config
import pandas as pd
import nltk

from datetime import date
```

```python
nltk.download('punkt')

user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent
```

```python
termos = ['Arcelormittal', 'Canadian solar','Aeroportos & Obras','Vale & investimento','Infraestrutura & Licitação',
         'Infraestrutura','CCR & Rodovia','Vale & Licitação','Aeroportos & Concessão',
         'Aeroportos & Fraport','Infraestrutura & Investimento']

df = pd.DataFrame(columns=['title','media','date','datetime','desc','link','img','termo'])

for termo in termos:

    googlenews=GoogleNews(period='1d',lang='pt')

    googlenews.search(termo)

    result=googlenews.result()
    
    result=pd.DataFrame(result)
    
    result['termo'] = termo
    
    df=pd.concat([df,result],ignore_index=True)
    

    
```

```python
df
```

```python
list=[]

for ind in df.index:
    dict={}
    article = Article(df['link'][ind],config=config)
    try:
        article.download()
        article.parse()
        article2 = article.text.split()
    except:
        print('***FAILED TO DOWNLOAD***', article.url)
        continue
    
    article.nlp()
    dict['Data']=df['datetime'][ind]
    dict['Termo']=df['termo'][ind]
    dict['Fonte']=df['media'][ind]
    dict['Titulo']=article.title
    dict['Descricao']=article.summary
    dict['URL']=article.url
    list.append(dict)
    
news_df=pd.DataFrame(list)
```
